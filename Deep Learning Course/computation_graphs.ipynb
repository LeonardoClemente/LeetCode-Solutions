```python
import numpy as np
```

```python

def sigmoid(x):
    return 1/(1 + np.exp(-x))

X = np.array([[1],[-1]])
W = np.array([[.5 ,.5], [0,1 ], [1,0]])
z = np.matmul(W,X)
h = sigmoid(z)
W_a = np.array([[-1], [.5], [.5]])
a = sigmoid(np.matmul(h.T, W_a))


y=1
```

```python
# Backprop
dL_da = -1*(y/a) + (1-y)*(1/(1-a))*-1 
da_dz = a*(1-a)
dz_h1 = W_a[0] 
dh1_dz1 =sigmoid(z[0])*(1-sigmoid(z[0]))
dz1_dw11 = X[0][0]
dz1_dw21 = X[1][0]

dL_dw11 = dL_da*da_dz*dz_h1*dh1_dz1*dz1_dw11
dL_dw21 = dL_da*da_dz*dz_h1*dh1_dz1*dz1_dw21


learning_rate = .5
new_weight = W[0][1] - learning_rate*dL_dw21
new_weight
```
